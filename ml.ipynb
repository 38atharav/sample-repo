```python
xpoints = np.array([1,2,5,15,20,13,75])
ypoints = np.array([5,10,100,200,63,42,300])
plt.plot(xpoints,ypoints, marker="o", ms=5, mec="b", color="pink")
plt.show()
```

```python
xpoints = np.array([1,2,50,15,20,10,75])
ypoints = np.array([5,10,100,200,63,42,300])
plt.plot(xpoints,ypoints, marker="*", ms=10, mec="b", color="pink")
plt.show()
```

```python
xpoints = np.array([1,2,5,15,20,13,75])
ypoints = np.array([5,10,100,200,63,42,300])
plt.plot(xpoints,ypoints, marker="^", ms=5, mec="b", color="pink",linestyle='dashed',linewidth=5)
plt.show()
```

```python
#mec=markr edge color ,mfc=marker face color,ms=mrkersize
plt.plot(xpoints,marker="o", ms=15, mec="b", color="pink",mfc='g',linestyle='dotted',linewidth=4)
plt.show()
```
